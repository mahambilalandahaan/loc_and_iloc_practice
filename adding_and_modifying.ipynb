```python
#Day 6: Adding & Modifying Columns (Tasks),Add a new column called "Age Group":,If Age < 18 → "Teen",If 18 ≤ Age < 40 → "Adult"',If Age ≥ 40 → "Senior"
#Add a new column called "Name Length" showing how many characters are in each student’s Name.
#Create a new column called "Email Provider" (just the part before @).
#Update the City name for one student (pick anyone) to a new value (e.g., change "Karachi" → "Islamabad").
from google.colab import drive
drive.mount('/content/Drive')
import pandas as pd
df=pd.read_csv('/content/Drive/My Drive/Colab Notebooks/students.csv')
#Clean and standardize data
df.rename(columns={"Full Name":"Name","Email address":"Email"},inplace=True)
df["Name"]=df["Name"].str.title().str.strip()
df["Email"]=df["Email"].str.lower().str.strip()
df["City"]=df["City"].str.title().str.strip()
#Adding a column "Age group"
df["Age group"]=pd.cut(df["Age"],bins=[0,18,40,100],labels=["Teen","Adult","Senior"])
#Adding a nere column age length
df["Name length"]=df["Name"].str.len()
#adding new column"Email provider
df["Email provider"]=df["Email"].str.split("@").str[0]
#update any student city name
df.loc[df["Name"]=="Atif","City"]="lahore"
print("UPDATED DATAFRAME\n ")
df.head()
```
